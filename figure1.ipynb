```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
raw = [76.3, 78.8, 79.8, 81.1, 82.6, 83.3, 84.0, 84.4, 85.0, 85.4] # b0-b7-b7RA-b8RA
raw_text = ['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B7RA', 'B8RA']
ns = [78.6, 81.3, 82.3, 84.0, 85.1, 86.0, 86.4, 86.9, 88.4] # b0-b7-l2
ns_text = ['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'L2']
fix = [80.2, 82.6, 83.6, 85.0, 85.9, 86.4, 86.7, 87.1, 88.5] # b0-b7-l2
fix_text = ['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'L2']
bit = [None, None, None, None, None, None, None, 85.3, 87.5]
ap = [77.0, 79.2, 80.3, 81.8, 83.2, 84.2, 84.7, 85.1, 85.3] # b0-b8
ap_text = ['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8']
```

```python
plt.figure(figsize=(10, 5))

plt.plot(fix, 'yo-', label='FixNet')
for i in range(len(fix)-1):
    plt.text(i, fix[i]+0.3, fix_text[i])
plt.text(8, fix[8]+0.2, fix_text[8])

plt.plot(ns, 'ro-', label='NoisyStudent')
for i in range(len(ns)):
    plt.text(i+0.1, ns[i]-0.3, ns_text[i])


plt.plot(bit, 'bo-', label='BigTransfer')
plt.text(7-0.4, bit[7]+0.3, 'BiT-M')
plt.text(8+0.1, bit[8]-0.2, 'BiT-L')


plt.plot(ap, 'go--', label='AdvProp')
for i in range(len(ap)):
    if i == 7:
        plt.text(i, ap[i]-0.45, ap_text[i])
    else:
        plt.text(i, ap[i]+0.35, ap_text[i])
        
plt.plot(raw, 'ko--', label='EfficientNet')
for i in range(len(raw)-2):
    plt.text(i+0.1, raw[i]-0.3, raw_text[i])
plt.text(8-0.1, raw[8]-0.6, raw_text[8])
plt.text(9-0.1, raw[9]-0.6, raw_text[9])


plt.xlabel('EfficientNet')
plt.ylabel('Imagenet Top-1 Accuracy (%)')
plt.xticks([i for i in range(10)], ['B0', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8/L2', 'B8/RA'])
plt.legend(loc='lower right', prop={'size': 16})
plt.savefig('fig1', dpi=1000)
```
